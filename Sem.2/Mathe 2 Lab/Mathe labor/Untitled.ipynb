```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_min, x_max, y_min, y_max = 8, 5, -1.0, 2.5
```

```python
fig, ax = plt.subplots(figsize=((x_max+x_min),(y_max+y_min)))
```

```python
ax.set_aspect('equal')
display(fig)
```

```python
ax.axis([x_min,x_max,y_min,y_max])
display(fig)
```
