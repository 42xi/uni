```python
import numpy as np
import matplotlib.pyplot as plt

# Pendulum parameters
m = 0.5  # kg
l = 0.8  # m
g = 9.82  # m/s^2
k = 0.1  # m*kg/s

# Initial conditions
phi_0 = np.pi / 4  # rad
phi_dot_0 = 0  # rad/s
t_end = 10  # s
dt = 0.01  # Time step

# Differential equation as a system of first-order ODEs
def pendulum_system(state, t):
    phi, phi_dot = state
    phi_ddot = -(k/(m*l)) * phi_dot - (g/l) * np.sin(phi)
    return np.array([phi_dot, phi_ddot])

# Time vector
t = np.arange(0, t_end + dt, dt)

# Initialize arrays for phi and phi_dot
state = np.zeros((len(t), 2))
state[0] = [phi_0, phi_dot_0]

# Euler method
for i in range(1, len(t)):
    state[i] = state[i-1] + dt * pendulum_system(state[i-1], t[i-1])

phi = state[:, 0]
phi_dot = state[:, 1]

# Plot phi(t)
plt.figure(figsize=(6, 4))
plt.plot(t, phi)
plt.title("Verlauf des Auslenkungswinkels $\\phi(t)$")
plt.xlabel("Zeit $t$ (s)")
plt.ylabel("Winkel $\\phi$ (rad)")
plt.tight_layout()
plt.show()

# Plot phase portrait phi_dot vs. phi
plt.figure(figsize=(6, 4))
plt.plot(phi, phi_dot)
plt.title("Phasenporträt $\\dot{\\phi}$ über $\\phi$")
plt.xlabel("Winkel $\\phi$ (rad)")
plt.ylabel("Winkelgeschwindigkeit $\\dot{\\phi}$ (rad/s)")
plt.tight_layout()
plt.show()
```
