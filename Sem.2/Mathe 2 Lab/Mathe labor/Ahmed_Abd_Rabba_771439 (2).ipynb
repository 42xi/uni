```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

x_min, x_max, y_min, y_max = -3, 3.5,-1,5.5
fig, ax = plt.subplots()

ax.axis([x_min,x_max,y_min, y_max])
ax.set_aspect('equal')
ax.grid(True)

ax.set_xticks(np.arange(x_min, x_max,1))
ax.set_yticks(np.arange(y_min, y_max, 1))

# Gitterlinien fuer Intervall 0.5
ax.set_xticks(np.arange(x_min, x_max+0.5, 0.5), minor=True)
ax.set_yticks(np.arange(y_min, y_max+0.5, 0.5), minor=True)

ax.grid(which='minor', alpha=0.3)
ax.grid(which='major', alpha=0.8)
```

```python
x=np.array([0,0,1,2,2,0,2,0,2])
y=np.array([0,2,4,2,0,2,2,0,0])
ax.plot(x,y,'k')
display(fig)
```

```python
grad=55*np.pi / 180
R=np.array([[np.cos(grad),-np.sin(grad)],[np.sin(grad),np.cos(grad)]])

X = np.array([x,y])
XR = R@X
xr = XR[0,:]
yr = XR[1,:]

ax.plot(xr,yr,'-r')
display(fig)
```
