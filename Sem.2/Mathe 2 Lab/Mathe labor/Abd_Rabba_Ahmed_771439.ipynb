```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
print(sp.__version__)
```

```python

x_0 = 0.0
dotx_0 = 0.25

a = 0.1
t0 = 0
t1 = 100.0
```

```python
def heartrateOde(t,z):
    zp = np.array([z[1], -a*(z[0]**2-1)*z[1] - z[0]])
    return zp
```

```python
def heartrateEuler(a):
    n = int((t1-t0)/a)
    t = np.linspace(t0,t1,n+1)
    z = np.zeros([len(z0),n+1])
    z[:,0] = z0
    for k in range(n):
        z[:,k+1] = z[:,k]+a*heartrateOde(t[k],z[:,k])
    return t, z
```

```python
z0 = np.array([x_0, dotx_0])
t, z = heartrateEuler(a)


fig, ax1 = plt.subplots()
ax1.plot(t,z[0,:],'-b')
ax1.set(xlabel='$t$', ylabel='$x$')
```

```python
fig, ax3 = plt.subplots()
ax3.plot(z[0,:],z[1,:],'-b')
ax3.set(xlabel='$x$', ylabel="\u0078\u0307") # \u0078\u0307 = x mit Punkt oben drauf
```
