```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import IFrame
```

```python
x_min, x_max, y_min, y_max = 0.0, 6.0, -1.0, 2.0 
fig, ax = plt.subplots(figsize=((x_max-x_min),(y_max-y_min)))
```

```python
ax.axis([x_min,x_max,y_min,y_max])
ax.grid(True)
```

```python
x_ticks = np.arange(x_min,x_max+1)
ax.set_xticks(x_ticks)
y_ticks = np.arange(y_min,y_max+1)
ax.set_yticks(y_ticks)
display(fig)
```

```python
ax.set_xticks(np.arange(x_min,x_max+1, 0.5),minor=True) 
ax.set_yticks(np.arange(y_min,y_max+1, 0.5),minor=True)
ax.grid(which='minor', alpha=0.1)
ax.grid(which='major', alpha=0.5)
display(fig)
```

```python
ax.scatter([1, 2, 3, 4, 5], [0, 0, 1, 0, 0], color='red', zorder=5, marker = "*")
display(fig)
```

```python
stars = np.array([1, 2, 3, 4, 5]) # x-Koordinaten
B = np.transpose(np.array([stars**0, stars**1, stars**2, stars**3, stars**4])) 

A = [0, 0, 1, 0, 0]  # y-Koordinaten

ergebnis = np.linalg.solve(B, A)

x_points = np.linspace(-16, 15,1000)
y_points = ergebnis[0] + ergebnis[1]*x_points + ergebnis[2]*x_points**2 + ergebnis[3]*x_points**3 + ergebnis[4]*x_points**4 
ax.plot(x_points, y_points, 'b')
display(fig)
```
