```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import IFrame
```

```python
x_min, x_max, y_min, y_max = -5, 5, -5, 5
fig, ax = plt.subplots(figsize=((x_max-x_min)/2,(y_max-y_min)/2))
```

```python
ax.axis([x_min,x_max,y_min,y_max])
display(fig)
```

```python
x_ticks = np.arange(x_min,x_max+1, 1)
ax.set_xticks(x_ticks)

y_ticks = np.arange(y_min,y_max+1, 1)
ax.set_yticks(y_ticks)

display(fig)
```

```python
ax.set_xticks(np.arange(x_min,x_max+0.5,0.5),minor=True)
ax.set_yticks(np.arange(y_min,y_max+0.5,0.5),minor=True)
ax.grid(which='minor', alpha=0.1)
ax.grid(which='major', alpha=0.5)
display(fig)
```

```python
t = np.linspace(0, 8*np.pi, 1000)  # Winkelwert in [0, 8*pi]
radius = np.sqrt(t)  # Radius hier halb so gro wie der Winkel

x = radius*np.cos(t)  # x-Koordinaten berechnen
y = radius*np.sin(t)  # y-Koordinaten berechnen
ax.plot(x,y, 'k')

display(fig)
```
