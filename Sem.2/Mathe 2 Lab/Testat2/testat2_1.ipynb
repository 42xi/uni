```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Sympy-Berechnungen
t = sp.symbols('t')
x = sp.Function('x')('t')

xp = sp.Derivative(x)
xpp = sp.Derivative(xp)

equation = sp.Eq(xpp + 3*xp + 2*x, 2*sp.sin(t))

konstanten = {x.subs(t, 0): 0, xp.subs(t, 0): 1}
awp = sp.dsolve(equation, ics=konstanten)

# Plot mit Matplotlib
t_vals = np.linspace(0, 10, 500)
awp_func = sp.lambdify(t, awp.rhs, 'numpy')
x_vals = awp_func(t_vals)

# Plot-Einstellungen
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(t_vals, x_vals, label='$x(t)$', color='blue', linewidth=2)

# Achsen wie im Bild
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

# Ticks und Labels
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Achsenbeschriftungen an den Enden
ax.set_xlabel('$t$', loc='right', fontsize=12)
ax.set_ylabel('$x(t)$', loc='top', fontsize=12)

# Gitter und Stil




plt.show()
```
