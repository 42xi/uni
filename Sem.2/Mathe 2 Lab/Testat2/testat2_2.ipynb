```python
import numpy as np
import matplotlib.pyplot as plt

B = 1.0
L0 = 0.5
h = 0.001
t0, tn = 0.0, 10.0
x0, dot_x0 = 0.6, 0.0

# DGL 2ter Ordnung
def fOde(t, z):
    x, x_dot = z
    x_ddot = -100 * (x_dot * x**2 - B * x_dot + x - L0)
    return np.array([x_dot, x_ddot])

def fEuler(t0, tn, h, z0):
    n = int((tn - t0) / h)
    t = np.linspace(t0, tn, n + 1)
    z = np.zeros((len(z0), n + 1))
    z[:, 0] = z0
    for k in range(n):
        z[:, k + 1] = z[:, k] + h * fOde(t[k], z[:, k])
    return t, z

# Initialwerte und Lösung berechnen
z0 = np.array([x0, dot_x0])
t, z = fEuler(t0, tn, h, z0)

fig, ax1 = plt.subplots()
ax1.plot(t, z[0, :])
ax1.set(xlabel='$t$', ylabel='$x(t)$')

fig, ax2 = plt.subplots()
ax2.plot(z[0, :], z[1, :])
ax2.set(xlabel='$x$', ylabel='$\dot{x}$')

plt.show()
```
